```python
import pandas as pd

df = pd.read_csv('/content/swadesh_list.csv')
```

```python
df.head()
```

```python
df = df.apply(lambda x: x.str.split(',').str[0] if x.dtype == 'O' else x)
```

```python
df = df.apply(lambda x: x.str.split('(').str[0] if x.dtype == 'O' else x)
```

```python
df.head()
```

```python
from unidecode import unidecode

df = df.apply(lambda x: x.apply(lambda y: unidecode(str(y)) if pd.notna(y) else y))
```

```python
df = df.apply(lambda x: x.str.replace('*', '') if x.dtype == 'O' else x)
```

```python
df.head()
```

```python
df = df.drop('№', axis=1)
```

```python
df.columns = df.columns.str.split().str[0]
```

```python
df.head()
```

```python
df.to_csv('/content/swadesh_list_clean.csv', index=False)
```
